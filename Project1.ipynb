# Analysis of Fradulent Credit Card Transaction Dataset

The dataset contains data of transactions made by Credit Cards in September 2013 by European cardholders.<br>
This dataset presents transactions that occurred in two days.<br>
The number of records in dataset is 284807.

Number of variables available in dataset is 31.

```python
import pandas as pd

data=pd.read_csv('creditcard.csv',nrows=297500)
print(data.head())
```

The below command calculated the mean, standard deviation, minimum value, 1st quartile, 2nd quartile (median), 3rd quartile and maximum value for all variables of dataset

```python
print(data.describe())
```

The below code prints the column names of the dataset

```python
#It provides column names (features) of dataset
for col in data.columns:
    print(col)
```

The below code checks whether there are any null values in each of the columns in the dataset.
If for a column, it is mentioned as not-null, it suggests that there are no null values in that column in the dataset.

```python
#to check whether columns have null values or not
print(data.info())
```

The below code generates plot for the data present in the column 'Class'.<br>
The plot shows that almost all transactions are valid transactions thus suggesting the dataset to be one-sided.<br>
Almost all transactions which are classified as '0' are non-fradulent.

```python
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [10, 6]
%config InlineBackend.figure_format = 'retina'

#label=['non_Fradulent','fraudulent']
data['Class'].value_counts().plot(kind="bar")

plt.xlabel('Class: 0=Non-fraudulent transactions, 1=Fraudulent Transactions')
plt.ylabel('Count')
plt.legend(loc="upper right")
plt.show()
```

Below code gives us the scatter plot between 'Time' and 'Amount'.<br>
We see some outliers in the plot and most of the amounts are below 5000<br>

```python
import matplotlib.pyplot as plt

plt.plot(data['Time'],data['Amount'], '+');
plt.xlabel('Time')
plt.ylabel('Amount')
plt.show()
```

In the code below, we are determining the points which have amount greater than 5000.<br>
We will be considering values which are marked as '1' as outliers.

```python
plt.plot(data['Time'],data['Amount'] > 5000,'+')
plt.show()
```

```python
import statsmodels.api as sm

estimators=['V1','V2','V3','V4','V5','V6','V7','V8','V9','V10','V11','V12','V13','V14','V15','V16','V17','V18','V19','V20',
            'V21','V22','V23','V24','V25','V26','V27','V28','Amount']
X1 = data[estimators]
y = data['Class']
col=X1.columns[:-1]
X = sm.add_constant(X1)
reg_logit = sm.Logit(y,X)
results_logit = reg_logit.fit()
```

```python
print(results_logit.summary())
```
